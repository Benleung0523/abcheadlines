```python
mental['What is your age?'].fillna(mental['What is your age?'].median(), inplace = True) #fill the missing age with median

def convert_to_male_or_female(val):   # convert the answer of "What is your gender?" into 3 unique answers only
    male = [
 'Cis Male',
 'Cis male',
 'Dude',
 "I'm a man why didn't you make this a drop down question. You should of asked sex? And I would of answered yes please. Seriously how much text can this take? ",
 'M',
 'MALE',
 'Male',
 'Male ',
 'Male (cis)',
 'Male (trans, FtM)',
 'Male.',
 'Male/genderqueer',
 'Malr',
 'Man',
 'M|',
 'Sex is male',
 'cis male',
 'cis man',
 'cisdude',
 'm',
 'mail',
 'male',
 'male ',
 'male 9:1 female, roughly',
 'man']
    female = [' Female','AFAB','Cis female ','Cis-woman','Cisgender Female','F',
 'Female',
 'Female ',
 'Female (props for making this a freeform field, though)',
 'Female assigned at birth ',
 'Female or Multi-Gender Femme','I identify as female.','Transgender woman','Woman',
'f',
 'fem',
 'female',
 'female ',
 'female-bodied; no feelings about gender',
 'female/woman',
 'fm','genderqueer woman','woman']
    if val in male:
        return 'male'
    if val in female:
        return 'female'
    return 'other/unspecified'
    
mental["What is your gender?"] = mental["What is your gender?"].apply(convert_to_male_or_female)
mental.fillna('other/unspecified', inplace = True)

```

## Analysis

### The frequency of mental health illness varies by geographic location:

```python
#filtering out those countries only have few respondents because it affects the normaalisation
#for example, if a single country that have only one respondent and that respondent has mental health disorder, it makes that country has 100% people of mental health disorder
countries=mental.loc[:,["Do you currently have a mental health disorder?","What country do you live in?"]]
countries = countries.groupby(["What country do you live in?"]).size().reset_index(name ='count')
countries = countries[countries['count']>= 10]
countries #the group of countries which have number of respondents >= 10
```

```python
frequency_vs_location =mental.loc[:,["Do you currently have a mental health disorder?","What country do you live in?"]]
frequency_vs_location =frequency_vs_location.groupby(["What country do you live in?","Do you currently have a mental health disorder?"]).size().reset_index(name ='count')
frequency_vs_location
```

```python
frequency_vs_location_filterred=frequency_vs_location.loc[frequency_vs_location["What country do you live in?"].isin(countries["What country do you live in?"])] #filtered with #the group of countries which have number of respondents >= 10
frequency_vs_location_filterred.head() #filtered by the group of countries which have number of respondents >= 10
```

```python
a = frequency_vs_location_filterred.groupby(["What country do you live in?"])["count"].transform('sum')
frequency_vs_location_filterred['normalised_count'] = frequency_vs_location_filterred['count'].div(a) #the counts are normalised by the sum of respondents of each countries
frequency_vs_location_filterred.head()
```

```python
frequency_vs_location_filterred_pivot=frequency_vs_location_filterred.pivot(index='Do you currently have a mental health disorder?', columns='What country do you live in?', values='normalised_count')
frequency_vs_location_filterred_pivot
```

### The attitudes towards mental health vary by geographic location

```python
attitudes_vs_location =mental.loc[:,["Would you feel comfortable discussing a mental health disorder with your coworkers?","What country do you live in?"]]
attitudes_vs_location =attitudes_vs_location.groupby(["What country do you live in?","Would you feel comfortable discussing a mental health disorder with your coworkers?"]).size().reset_index(name ='count')
attitudes_vs_location.head()
```

```python
attitudes_vs_location_filterred=attitudes_vs_location.loc[attitudes_vs_location["What country do you live in?"].isin(countries["What country do you live in?"])] #filtered with #the group of countries which have number of respondents >= 10
attitudes_vs_location_filterred.head() #filtered by the group of countries which have number of respondents >= 10
```

```python
a = attitudes_vs_location_filterred.groupby(["What country do you live in?"])["count"].transform('sum')
attitudes_vs_location_filterred['normalised_count'] = attitudes_vs_location_filterred['count'].div(a) #the counts are normalised by the sum of respondents of each countries
attitudes_vs_location_filterred.head()
```

```python
attitudes_vs_location_filterred_pivot=attitudes_vs_location_filterred.pivot(index='Would you feel comfortable discussing a mental health disorder with your coworkers?', columns='What country do you live in?', values='normalised_count')
attitudes_vs_location_filterred_pivot
```

### The strongest predictors of mental health illness

```python
labeled_mental = mental.applymap(str) 
labelDict = {} # give label to the features
for feature in labeled_mental:
    le = preprocessing.LabelEncoder()
    le.fit(labeled_mental[feature])
    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    labeled_mental[feature] = le.transform(labeled_mental[feature])
    # Get labels
    labelKey = 'label_' + feature
    labelValue = [*le_name_mapping]
    labelDict[labelKey] =labelValue
    
```

```python
corrmat = labeled_mental.corr()
corr_mdh = corrmat["Do you currently have a mental health disorder?"].sort_values(ascending=False).head(10)
corr_mdh
```

### The strongest predictors of specific attitudes towards mental health

```python
corr_atmh =corrmat["Would you have been willing to discuss a mental health issue with your previous co-workers?"].sort_values(ascending=False).head(10)
corr_atmh
```

# Question 3: What were the top Australian news topics over the last decade, and what can these say about the national conversation?

## Interpretion of this question：
**News topics** would be the subject of news ,for example, politics, technology,economy... etc.  
**The top Australian news topics** would be the news topics which oucurred the most over the last decade.  
**National conversation** would be the hot topics what people are talking about. Here the threads in Reddit would be chosen to represent national conversation.  
**"What can these say about"** means the relationship between the top Australian news topics and national conversation and how they affect each other.  
## Key stakeholders who have an interest in the question:
- **Government**: 

  Government can have a better understanding about the major concerns of citizens so that it can implement policies corresponding to those concerns.

- **Media industry**:

  By knowing what were the hot topics, media industry can produce more eye-catching contents about the hot topics and make more profits.
  
- **Research institutes of social science**:

  It is an important material for social science research.
  


```python


from sklearn.feature_extraction.text import CountVectorizer
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import MultinomialNB 
```

### Collect / Load Data

For this case study, I will make use of the following data:

- [A Million News Headlines](https://www.kaggle.com/therohk/million-headlines)
- [Worldnews on Reddit from 2008 to Today](https://www.kaggle.com/rootuser/worldnews-on-reddit)


```python

abc = pd.read_csv("abcnews-date-text.csv",parse_dates=[0], infer_datetime_format=True)
abc = abc[(abc['publish_date'] >= '2009-01-01')] #filterring the years not belong to last decade(2009-1019)
abc
```

```python

reddit = pd.read_csv("reddit_worldnews_start_to_2016-11-22.csv")
reddit= reddit.loc[:,["date_created","title"]]
reddit = reddit[(reddit['date_created'] >= '2009-01-01')] #filterring the years not belong to last decade(2009-1019)
reddit
```

## Analysis

### Top Australian news topics

```python
data = fetch_20newsgroups()  #trainning a text classifier
categories = data.target_names
train = fetch_20newsgroups(subset='train', categories=categories)
test = fetch_20newsgroups(subset='test', categories=categories)
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(train.data, train.target)

def predict_category(s, train=train, model=model):  #a function to classify the text
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

### Top news topics in national conversation

```python
abc_prediction['publish_date'] =  pd.to_datetime(abc_prediction['publish_date'])
abc_prediction_count = abc_prediction.groupby(["predictions"])['predictions'].count().reset_index(name='count')
abc_prediction_count = abc_prediction_count.sort_values('count', ascending=True)
abc_prediction_count
```

```python
reddit_predrction
```

```python
reddit_predrction['date_created'] =  pd.to_datetime(reddit_predrction['date_created'])
reddit_predrction_count = reddit_predrction.groupby(["predictions"])['predictions'].count().reset_index(name='count')
reddit_predrction_count = reddit_predrction_count.sort_values(['count'], ascending=True)
reddit_predrction_count
```

```python
abc_prediction_to2016=abc_prediction[(abc_prediction['publish_date'] <= '2016-11-22')] #to make the comparison because of the reddit dataset is from 2008 to 2016
abc_trend=abc_prediction_to2016.groupby([abc_prediction_to2016['publish_date'].dt.year,"predictions"])["predictions"].count().reset_index(name='count')
abc_trend_pivot = abc_trend.pivot(index='publish_date', columns='predictions', values='count')
abc_trend_pivot
```

```python
reddit_predrction
reddit_trend=reddit_predrction.groupby([reddit_predrction['date_created'].dt.year,"predictions"])["predictions"].count().reset_index(name='count')
reddit_trend_pivot = reddit_trend.pivot(index='date_created', columns='predictions', values='count')
reddit_trend_pivot
```
